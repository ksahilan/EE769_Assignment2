Removing the Outliers

```python
r_clean_data = red_wine.copy(deep=True) # to make a independent copy of the data, refered pandas documentation
    # refered https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.copy.html
    # Also refered Box and Whisker plot webpage mentioned previously
for col in X_r.columns:
    q1, q3 = np.percentile(r_clean_data[col], [25, 75])
    # Calculate the interquartile range
    iqr = q3 - q1
    # Calculate the lower and upper bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    # Drop the outliers
    r_clean_data = r_clean_data[(r_clean_data[col] >= lower_bound) 
                    & (r_clean_data[col] <= upper_bound)]
r_clean_data.to_csv('clean_red_wine.csv', sep=';')

```

```python
w_clean_data = white_wine.copy(deep=True) # to make a independent copy of the data, refered pandas documentation
    # refered https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.copy.html
    # Also refered Box and Whisker plot webpage mentioned previously
for col in X_w.columns:
    q1, q3 = np.percentile(w_clean_data[col], [25, 75])
    # Calculate the interquartile range
    iqr = q3 - q1
    # Calculate the lower and upper bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    # Drop the outliers
    w_clean_data = w_clean_data[(w_clean_data[col] >= lower_bound) 
                    & (w_clean_data[col] <= upper_bound)]
w_clean_data.to_csv('clean_white_wine.csv', sep=';')

```

Removing the Outliers using IQR

```python
X_c_cd = X_c.copy(deep=True) # to make a independent copy of the data, refered pandas documentation
    # refered https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.copy.html
    # Also refered Box and Whisker plot webpage mentioned previously
for col in X_c.columns:
    q1, q3 = np.percentile(X_c_cd[col], [25, 75])
    # Calculate the interquartile range
    iqr = q3 - q1
    # Calculate the lower and upper bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    # Drop the outliers
    X_c_cd = X_c_cd[(X_c_cd[col] >= lower_bound) 
                    & (X_c_cd[col] <= upper_bound)]
Y_c_cd = Y_c.loc[X_c_cd.index]
print("No. of Data points changed from", X_c.shape,"-->",X_c_cd.shape)
```

This is huge loss in data. Thus let us try outlier removal using z-score.

```python
import scipy.stats as stats
from scipy.stats import zscore
threshold = 3
for col in X_c.columns:
        # Drop the outliers
        X_c_cd_z = X_c[(np.abs(stats.zscore(X_c[col])) < threshold)]
Y_c_cd_z = Y_c.loc[X_c_cd_z.index] # to get the corresponding Y values for the clean X values
print("No. of Data points changed from", X_c.shape,"-->",X_c_cd_z.shape)
```

## Recursive Feature Elimination

```python
from sklearn.feature_selection import RFECV

def best_model(X_train, Y_train, model):
    rfecv = RFECV(estimator=model, step=1, cv=StratifiedKFold(shuffle=True, random_state=42), scoring='accuracy')
    rfecv = rfecv.fit(X_train, Y_train)
    print("Optimal number of features : %d" % rfecv.n_features_)
    print("Ranking of features: ", rfecv.ranking_)
    print("Feature Selection: ", rfecv.support_)
```
